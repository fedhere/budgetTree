```python
budget2012 = {"name": "2012",
             "children": []}
```

```python
budget2017 = {"name": "2017",
             "children": []}
```

```python
for i,dept in enumerate(budgetGroup.index.levels[0]):
    #if i>20: break
    budget2012["children"].append({"name":dept, "children":[]})
    for ag in budgetGroup.loc[dept].index.levels[0]:
        try: 
            if not ((budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2010"] == 0 ) + 
                    (budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2012"] == 0)):
                print dept, ag, "D", budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2012"] 
                budget2012["children"][i]["children"].append({"name":ag, 
                                                            "size":budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2012"] * 0.000001,
                                                             "oldsize":budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2010"] * 0.000001,
                                                            "kind":"Discretionary",
                                                             "diff": float(budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2012"] -
                                                              budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2010"]) / 
                                                              budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2010"]})
        except KeyError:
            continue
        try: 
            if not ((budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2010"] == 0 ) + 
                    (budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2012"] == 0)):
                print dept, ag, "D", budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2012"]
                budget2012["children"][i]["children"].append({"name":ag, 
                                                            "size":budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2012"] * 0.000001,
                                                             "oldsize":budgetGroup.loc[dept].loc[ag].loc["Discretionary"]["2010"] * 0.000001,
                                                            "kind":"Mandatory",
                                                             "diff": float(budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2012"]- \
                                                              budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2010"]) / 
                                                              budgetGroup.loc[dept].loc[ag].loc["Mandatory"]["2010"]})

        except KeyError:
            continue
```
